Read data from csv file stored in drive and load it as a dataframe.Drop rows if it has any null values. print top 5 rows. 

```
import pandas as pd
#load data
df = pd.read_csv("/content/drive/My Drive/covid.csv")
#Prints first five rows
df.dropna(axis=0,how='any',inplace=True)
df.head()
```

Check the type of each feature in dataset,here int and float is of numerical data and object is categorical data. 

```
df.info()
```

```
#MEASURE OF CENTER
#prints mean of the selected column
avg = df['Confirmed'].mean()
print("Confirmed cases mean is:",avg)
#prints median of the selected column
median = df['Confirmed'].median()
print("Confirmed cases median is:",median)
```

```
#MEASURE OF VARIATION
# To analyse the variation of the selected column in the dataset
standard_deviation = df['Confirmed'].std()
print("std of confirmed cases:",standard_deviation)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(style="whitegrid")
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Confirmed', data=df, orient="v")
```

```
sns.set(style="whitegrid")
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='WHO Region', y='Confirmed', data=df, orient="v")
```

```

plt.figure(figsize=(14,8))
sns.distplot(df['Confirmed'], kde=True)
```

```
type_counts = df['WHO Region'].value_counts()
df2 = pd.DataFrame({'WHO Region': type_counts}, 
                     index = ['Eastern Mediterranean','Europe','Africa','Americas','Western Pacific','South-East Asia']
                   )
df2.plot.pie(y='WHO Region', figsize=(10,10), autopct='%1.1f%%')
```

```
sns.set(style='darkgrid')
plt.figure(figsize=(14,8))
ax = sns.countplot(x='Country/Region', data=df.head())
```
